### Import Dataset

```python
import json
import pandas as pd

data = []

with open(r"C:\Users\acer\OneDrive\Desktop\College Work\SEM 6\FINAL PROJECT\data\yelp_academic_dataset_review.json", "r", encoding="utf-8") as f:
    for line in f:
        data.append(json.loads(line))

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
df.columns
```

### Only keep required columns

```python
df = df[['business_id', 'text', 'stars']]
df.columns = ['business_id', 'review_text', 'rating']
```

```python
df.head()
```

### Filter and reduce the dataset

```python
df = df.groupby('business_id').filter(lambda x: len(x) >= 10)
```

```python
df = df.sample(n=20000, random_state=42)
```

```python
df.to_csv("yelp_reviews.csv", index=False)
```
